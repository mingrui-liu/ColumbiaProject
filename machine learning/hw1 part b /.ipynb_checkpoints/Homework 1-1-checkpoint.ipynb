```python
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import pandas as pd
```

```python
# Load data
data = pd.read_csv("problem3.csv", header = None)
data = np.array(data)
```

```python
# Define log likelihood
# theta contains x0 and gamma
def negloglikelihood(theta, data):
    num_points = data.shape[0]
    x0, gamma = theta
    loglikelihood = num_points * np.log(np.pi) + num_points * gamma
    loglikelihood = loglikelihood + sum(np.log(1+((data-x0)/(np.exp(gamma))*(data-x0)/(np.exp(gamma)))))
    return loglikelihood
```

```python
fig = plt.figure(figsize=(8,4), dpi=200)
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-6, 6, 0.5)
y = np.arange(-6, 6, 0.5)
X, Y = np.meshgrid(x, y)

def fun(x, y, data):
    return np.array([negloglikelihood((x_value, y_value), data) for x_value, y_value in zip(x,y) ])
zs = np.array(fun(np.ravel(X), np.ravel(Y), data))
Z = zs.reshape(X.shape)
print(Z.shape, Y.shape, X.shape)

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=0.5)

ax.set_xlabel('x0')
ax.set_ylabel('gamma')
ax.set_zlabel('Negative Log Likelihood')

# Get current rotation angle
print (ax.azim, ax.elev)
 
# Set rotation angle to 30 degrees
ax.view_init(azim=300)
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
def negloglikelihood_grad(theta, data):
    num_points = data.shape[0]
    x0, gamma = theta
    grad = np.zeros((2,))

    # The partial with respect to x0.
    grad[0] = sum(2*((data-x0)/((np.exp(gamma))**2))/(1+((data-x0)/np.exp(gamma))**2))
  
    # The partial with respect to sigma.
    grad[1] = sum(2*((data-x0)/np.exp(gamma))**2/(1+((data-x0)/np.exp(gamma))**2)) - num_points

    return - grad
```

```python
# Initial iterate: the first is mu, the second is sigma.
theta = np.array([1.0, 2.0])

fval_history = []
theta_history = []
def objective_function(theta, data):
    fval = negloglikelihood(theta, data)
    global fval_history
    fval_history += [ fval ]
    global theta_history
    theta_history += [ theta ]
    return fval

def objective_gradient(theta, data):
    gradient = negloglikelihood_grad(theta, data)
    return gradient

res = minimize(objective_function, theta, method = "BFGS", jac=objective_gradient, args=(data,))
```

```python
def GDRidge(x, y, T, eta, l):
    num_col = x.shape[1]
    n = len(y)
    b = 0
    w = np.zeros(num_col)
    log_w = []
    log_L = []
    for i in range(T):
        log_w += [weight_norm(w)]
        log_L += [reg_obj(x, y, w, b, l)]
        s_b = 0
        s_w = 0
        for j in range(n):
            s_b += 2*(y[j]-(b+np.dot(w,x[j])))
            s_w += 2*(y[j]-(b+np.dot(w,x[j])))*x[j]
        b_new = b + 1/n*s_b*eta
        w_new = w + eta*(-2*l*w + s_w/n)
        b = b_new
        w = w_new
    return b, w, log_L, log_w
```

### c

```python
from sklearn.datasets import load_boston
from sklearn import preprocessing
import numpy as np
boston = load_boston()
X = boston.data
y = boston.target
feature = boston.feature_names
```

```python
X = preprocessing.scale(X)
y = y.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
y = min_max_scaler.fit_transform(y)
y = y.reshape(-1)
```

```python
def reg_obj(x, y, w, b, l):
    n = len(y)
    s = 0
    for j in range(n):
        s += (y[j]-(b+np.dot(w,x[j])))**2
    result = 1/n*s + l*np.dot(w, w)
    return result

def weight_norm(w):
    return np.sqrt(np.dot(w, w))
```

```python
b, w, log_L, log_w = GDRidge(X, y, T = 100, eta = 0.01, l = 0.1)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(range(100), log_L, s = 9, label = "Reg Objective Value")
plt.scatter(range(100), log_w, s = 9, label = "2-norm of the weight")
plt.legend()
```

### d

```python
lam = np.array([0.001, 0.01, 0.1, 1, 10])
num_col = X.shape[1]
df_w = []
for i in lam:
    b, w, log_L, log_w = GDRidge(X, y, T = 100, eta = 0.01, l = i)
    df_w.append(w)

df_w = np.transpose(np.array(df_w))

plt.figure(figsize=(15,7))
for j in range(num_col):
    plt.plot(lam, df_w[j], label = feature[j])
plt.legend()
plt.xlabel("$\lambda$")
plt.ylabel("coefficient")
plt.xscale("log")
```

The coefficients become closer to 0 as the regularization gets bigger.

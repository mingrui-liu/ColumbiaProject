##### Negative Log Likelihood

```python
def neg_log_likelihood (theta,df):
    n = df.shape[0]
    x0,gamma = theta
    
    loglikelihood = -n*np.log(np.pi)-n*gamma-sum(np.log(1+((df-x0)/np.exp(gamma))**2))
    return -loglikelihood
```

```python
fig = plt.figure(figsize=(8,4), dpi=200)
ax = fig.add_subplot(111, projection='3d')
x = np.arange(-6, 6, 0.5)
y = np.arange(-6, 6, 0.5)
X, Y = np.meshgrid(x, y)

def fun(x, y, data):
    return np.array([neg_log_likelihood((x_value, y_value), data) for x_value, y_value in zip(x,y) ])

zs = np.array(fun(np.ravel(X), np.ravel(Y), df))
Z = zs.reshape(X.shape)
print(Z.shape, Y.shape, X.shape)

ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=0.5)

ax.set_xlabel('x0')
ax.set_ylabel('gamma')
ax.set_zlabel('Negative Log Likelihood')

# Get current rotation angle
print (ax.azim, ax.elev)
 
# Set rotation angle to 30 degrees
ax.view_init(azim=300)
plt.show()
```

### (d)
##### Gradient of Negative Log Likelihood

```python
# Negative log likelihood function gradient (consists of partial derivatives).
# The first component is the partial with respect to x0.
# The second component is the partial with respect to gamma.
def neg_loglikelihood_grad(theta, data):
  n = data.shape[0]  
  x0,gamma = theta
  grad = np.zeros((2,))

  # The partial with respect to x0.
  grad[0] = 2/np.exp(gamma)*sum(2*((data-x0)/((np.exp(gamma))**2))/(1+((data-x0)/np.exp(gamma))**2))
  
  # The partial with respect to gamma.
  grad[1] = -n + sum(2*((data-x0)/np.exp(gamma))**2/(1+((data-x0)/np.exp(gamma))**2)) 

  return - grad
```

```python
# Initial iterate: the first is x0, the second is gamma.
theta = np.array([0.0, 10.0])

fval_history = []
theta_history = []
def objective_function(theta, data):
    fval = neg_log_likelihood(theta, data)
    global fval_history
    fval_history += [ fval ]
    global theta_history
    theta_history += [ theta ]
    return fval

def objective_gradient(theta, data):
    gradient = neg_loglikelihood_grad(theta, data)
    return gradient

res = minimize(objective_function, theta, method = "BFGS", jac=objective_gradient, args=(df,))
print(res)
```

```python
result = {"fval":fval_history,"theta": theta_history}
pd.DataFrame(result)
```

```python
def objective_function(w,b,x,y,Lambda):
    n = len(y)
    s = 0
    for j in range(n):
        s += (y[j]-(b+np.dot(w,x[j])))**2
        
    result = 1/n*s + Lambda*np.dot(w, w)
    return result
```

```python
def GDRidge(x,y,T,eta,Lambda):
    b = 0
    w = np.zeros(x.shape[1])
    log_w = []
    log_L = []
    n = len(y)

    for t in range(T):
        log_w.append(np.sqrt(sum(w**2)))
        log_L.append(objective_function(w,b,x, y,Lambda))
        s_b = 0
        s_w = 0

        for j in range(n):
            s_b += 2*(y[j]-(b+np.dot(w,x[j])))
            s_w += 2*(y[j]-(b+np.dot(w,x[j])))*x[j]
        b_new = b + 1/n*s_b*eta
        w_new = w + eta*(-2*Lambda*w + s_w/n)
        b = b_new
        w = w_new
    return b, w, log_L, log_w
```

```python
from sklearn.datasets import load_boston
from sklearn import preprocessing

boston = load_boston()
X = boston.data
y = boston.target
feature = boston.feature_names
```

```python
X = preprocessing.scale(X)
y = y.reshape(-1,1)
min_max_scaler = preprocessing.MinMaxScaler()
y = min_max_scaler.fit_transform(y)
y = y.reshape(-1)
```

```python
b, w, log_L, log_w = GDRidge(X, y, T = 100, eta = 0.01, Lambda = 0.1)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(range(100), log_L, s = 9, label = "Reg Objective Value")
plt.scatter(range(100), log_w, s = 9, label = "2-norm of the weight")
plt.legend()
```

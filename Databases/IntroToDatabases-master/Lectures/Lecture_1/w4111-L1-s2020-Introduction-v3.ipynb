#### How Many Rows are in the Files (Database)

````python
# Import some helpful libraries.
import pandas as pd
import time
import os
````

````python
#
# Let's look at the files in the directory of information.
#
file_dir = "../../Data/IMDB/TSVFormat"
data_files = os.listdir(file_dir)

print("The list of data files are:")
for f in data_files:
    print("\t", f)
````

- These are like spreadsheet/Excel/Pages/Google Sheets files.


- There is a Python library, like in all languages, for reading these files. The following function simply opens the files and counts the number of rows.

````python
import csv

def count_rows(fn, delimiter):
    # fn is the fully qualified path and file name.
    # Delimiter is what separates the columns in the "spreadsheet."
    
    result = 0
    print("FN = ", fn)
    
    # Standard approach to opening a text file.
    with open(fn, "r") as in_file:
        
        # Use the CSV Reader to simplify parsing the the text file.
        # Delimiter specifies what separates the columns.
        # There are several other options, including whether or not quotes characters have a role.
        csv_rdr = csv.DictReader(in_file, delimiter=delimiter, quoting=csv.QUOTE_NONE)
        
        # For each row in the file.
        for r in csv_rdr:
            try:
                if result == 0:
                    print("For file = ", fn)
                    print("The colums are: ", csv_rdr.fieldnames)
                    print("The first rows as a dict is: ", r)
                result = result + 1
            except Exception as e:
                print("Could not read row, Exception = ", e)
                print("r = ", r)
                print("row count = ", result)

    return result
````

- Just for fun, let's see how much time it takes to read all of the files?

````python
total_rows = 0

start_time = time.time()
for f in data_files:
    print("\n**********")
    result = count_rows(file_dir + "/" + f, delimiter="\t")
    total_rows += result
end_time = time.time()
elapsed_time = end_time-start_time

print("\n\nThe total number of rows is = ", total_rows)
print("Total time to do the reading is ...", elapsed_time)
````

    - There are 85 million rows.
    - Reading all of them took 5 minutes.

````python
def load_csv_file(fn, delimiter="\t", quoting=csv.QUOTE_NONE):
    """
    Reads a CSV file and returns a list of dictionaries representing each row.

    :param fn: Fully qualified path to the file.
    :param delimiter: Field delimiter.
    :param quoting: Quoting value for strings.
    :return: Array of OrderedDict representing rows in file.
    """

    result = []

    with open(fn, "r") as in_file:

        # Create a DictReader with the options passed into this function.
        csv_rdr = csv.DictReader(in_file, delimiter=delimiter, quoting=quoting)
        for r in csv_rdr:
            result.append(dict(r))

    return result
````

````python


start_time = time.time()

name_basics_tbl = load_csv_file("../../Data/IMDB/TSVFormat/name_basics.tsv")
load1_time = time.time() - start_time
print("Loaded ", len(name_basics_tbl), 'name_basics rows in ', load1_time, 'seconds.')

title_basics_tbl = load_csv_file("../../Data/IMDB/TSVFormat/title_basics.tsv")
load2_time = time.time() - start_time
print("Loaded ", len(title_basics_tbl), 'title_basics rows at elapsed time ', load2_time, 'seconds.')

title_principals_tbl = load_csv_file("../../Data/IMDB/TSVFormat/title_principals.tsv")
load3_time = time.time() - start_time
print("Loaded ", len(title_principals_tbl), 'title_principals rows at elapsed time ', load3_time, 'seconds.')

````

- Let's write a generic function that will find entries in a loaded TSV table by field and value.

````python
def find_by_field(csv_table, field_name, value):

    result = []

    for r in csv_table:
        test_v = r.get(field_name, None)
        if test_v == value:
            result.append(r)

    return result
````

##### Structured Data

- High degree of organization.


- All records for a given type have
    - The same fields.
    - The fields are strongly typed.
    - There are _integrity constraints_ on values, e.g. __NOT NULL.__
    
    
- The _relational model_ is the canonical example.
    - The data is __structured,__ which means you must define the structure $\Rightarrow$ there is a __data definition language (DDL).__ A DDL is similar to defining a class in Java, C++ or Python.
    - Instances of data (rows, tuples) must match the DDL in the same way objects match the class definition.
    - The Relational Model has additional capabilities:
        - Integrity constraints, e.g. ```NOT NULL.``` _Primary Key_ means that the $student\_id$ is ```NOT NULL``` and ```UNIQUE.```
        - Referential Integrity:  For every _Grade_ row, there must be one row with a matching $student\_id$ in _Students._
        - Indexes: There is an auxiliary data structure that speeds up locating students using name fields.

##### Semi-structured Data

- There are records with fields.


- You can parse the data to extract records and fields into data structures.


- Records may have some common fields, but any record may have any fields of any types.


- Documents are the canonical example:
    - There are common structural concepts: headers, lists, footnotes, etc.
    - But different documents have different combinations, nesting, etc.
    
    
- [CouchDB](http://couchdb.apache.org/) is an example of a document database. It stores semi-structured [JSON](https://www.json.org/) documents.

#### [Data versus Metadata](https://en.wikipedia.org/wiki/Metadata)

##### Concept

- Metadata is "data information that provides information about other data". Three distinct types of metadata exist:
    1. Descriptive metadata
    2. Structural metadata
    3. Administrative metadata.
    
    
- Descriptive metadata describes a resource for purposes such as discovery and identification. It can include elements such as title, abstract, author, and keywords.


- Structural metadata is metadata about containers of data and indicates how compound objects are put together, for example, how pages are ordered to form chapters. It describes the types, versions, relationships and other characteristics of digital materials.


- Administrative metadata provides information to help manage a resource, such as when and how it was created, file type and other technical information, and who can access it.


- We will cover metadata, especially while studying relational data.
